```python
import os
import sys
import csv
import glob
import numpy as np
import matplotlib.pyplot as plt
```

```python
def calc_decomp_freqs(STATE_DECOMP_FILE):

	"""
		Parse state decomposition file and calculate frequencies
		for active, intermediate 1,2,3 and inactive state. 
	"""
	f = open(STATE_DECOMP_FILE, 'r')
	state_freq = np.array([0]*5)
	for line in f:
		state = int(line.strip().split("\t")[1])
		# print(state)
		state_freq[state-1] += 1
	total_frames = sum(state_freq)
	state_freq = state_freq/float(total_frames)
	return state_freq
```

```python
p = glob.glob("/scratch/PI/rondror/akma327/DynamicNetworks/data/DynamicNetworksOutput/InteractionOutputDictionary/*/*/rep*/0505*/state_classification.txt")
for a in sorted(p):
    print (":".join(a.split("/")[9:12]))
    if("inactive" in a):
        print(calc_decomp_freqs(a)[0:2])
    else:
        print(calc_decomp_freqs(a)[3:5])
    print ""
```
